```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('train.csv')
train
```

```python
# Histogram
sns.histplot(train['Age'], bins=30, kde=True)#bins=30: Defines the number of intervals in the histogram
#kde=True: Adds a smooth density curve to highlight patterns.
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(train['Fare'], bins=40, kde=True)
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(train, x="Age", hue="Survived", bins=30, kde=True)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot
sns.boxplot(x=train['Age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=train)
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=train)
plt.title("Age Distribution Across Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
sns.boxplot(x=train['Fare'])
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=train)
plt.title("Age vs. Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.scatterplot(x='SibSp', y='Parch', data=train, hue='Survived')
plt.title("Siblings/Spouses vs. Parents/Children")
plt.xlabel("SibSp")
plt.ylabel("Parch")
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train)
plt.title("Fare vs. Age Colored by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
